```python
import networkit as nk
from plots import Plots
from tools import values_frequency
```

```python
snapReader = nk.graphio.SNAPGraphReader(True, False)
G = snapReader.read("data/email-Eu/email-Eu-core.txt")
```

```python
in_degrees = [G.degreeIn(v) for v in G.iterNodes()]
in_degrees, freq = values_frequency(in_degrees)
Plots.degree_distribution(in_degrees, freq, "In-degrees")

out_degrees = [G.degreeOut(v) for v in G.iterNodes()]
out_degrees, freq = values_frequency(out_degrees)
Plots.degree_distribution(out_degrees, freq, "Out-degrees")
```
